# $\alpha$ variance

````python
import pandas as pd
import numpy as np
import pickle as pkl
import warnings
from tqdm import tqdm
from scipy import optimize
from fisher import pvalue
from multiprocessing import Pool
from plotnine import *
import subprocess
import glob
import sys
import os
sys.path.insert(0, '/home/jmurga/mkt/201902/scripts/src/')
from mkt import *

def dofe_to_sfs(df):

    n = df.iloc[:,1].values[0]

    pn = df.iloc[:,3:(n+2)].to_numpy().flatten()
    ps = df.iloc[:,(n+3):(n+n+2)].to_numpy().flatten()
    m = df.iloc[:,n+n+2:].to_numpy().flatten()
    d = m[[1,3]]

    sfs = pd.DataFrame({'f':np.arange(1,pn.shape[0]+1)/pn.shape[0],'pn':pn,'ps':ps}).to_numpy()

    checkDaf = sfs[sfs[:,0] > 0.25,:]
    if((np.sum(checkDaf[:,1]) == 0) | (np.sum(checkDaf[:,2]) == 0)):
        imp1 = {"alpha":np.nan,"pvalue":np.nan}
    else:
        imp1 = impMK(sfs=sfs,divergence=d,m=m,l=0.25);
    std = standardMK(sfs,d,m)
    return(np.array([std['alpha'],imp1['alpha']]))


out    = []
grapes = []
for n in glob.glob(PATH + "/raw_data/dofe/pooled/zi/*"):
    tmp_grapes = pd.read_csv(n + "/test",names = ['id','model','grapes','alpha_down','alpha_up','ci']).iloc[:,[0,2]]
    tmp_grapes['bins'] = n.split('/')[-1]
    grapes.append(tmp_grapes.iloc[:,[0,2,1]])

    for i in tqdm(glob.glob(n + "/*.dofe")):
        tmp = pd.read_csv(i,sep='\t',skiprows=2,header=None)
        y = dofe_to_sfs(tmp)
        b,f = i.split('/')[-2:]
        out.append(pd.DataFrame({'id':f.split('.')[0],'bins':b,'std':y[0],'imp':y[1]},index=[0]))

x      = pd.concat(out)
grapes = pd.concat(grapes)
y = pd.merge(x,grapes)

y.bins = y.bins.astype(float)
y = y.sort_values(['bins'])

c = []
s = []
v = []
for i in y.bins.unique():                       
    tmp = y[y.bins==i]                          
    tmp_c = tmp.iloc[:,2:].dropna().corr()
    tmp_s = tmp.iloc[:,2:].dropna().std()
    tmp_v = tmp.iloc[:,2:].dropna().var()

    c.append(tmp_c.to_numpy())
    s.append(tmp_s.to_numpy())
    v.append(tmp_v.to_numpy())

c = pd.DataFrame(np.vstack(c),columns=['mkt','imp'])
s = pd.DataFrame(np.vstack(s),columns=['mkt','imp','grapes'])
v = pd.DataFrame(np.vstack(v),columns=['mkt','imp','grapes'])

v['bins'] = y.bins.unique()
````
