```r
library(data.table)
library(iMKT)
setwd('/home/jmurga/mkt/201902/rawData/simulations/wdEstimation/neutral1')
listDaf1 <- lapply(list.files(pattern = 'daf',recursive=T),read.table,sep='\t',header=T)
listDiv1 <- lapply(list.files(pattern = 'div',recursive=T),read.table,sep='\t',header=T)
setwd('/home/jmurga/mkt/201902/rawData/simulations/wdEstimation/neutral2')
listDaf2 <- lapply(list.files(pattern = 'daf',recursive=T),read.table,sep='\t',header=T)
listDiv2 <- lapply(list.files(pattern = 'div',recursive=T),read.table,sep='\t',header=T)
setwd('/home/jmurga/mkt/201902/rawData/simulations/wdEstimation/neutral3')
listDaf3 <- lapply(list.files(pattern = 'daf',recursive=T),read.table,sep='\t',header=T)
listDiv3 <- lapply(list.files(pattern = 'div',recursive=T),read.table,sep='\t',header=T)
```

```r
m1 <- matrix(nrow = 2,ncol=length(listDaf))
m2 <- matrix(nrow = 2,ncol=length(listDaf))
m3 <- matrix(nrow = 2,ncol=length(listDaf))

for(i in 1:length(listDaf))
{
    print(i)
    r1 <- aMKT(daf=listDaf1[[i]][,1:3],divergence=listDiv1[[i]][,1:4],xlow=0,xhigh=1,plot=F)
    r2 <- aMKT(daf=listDaf2[[i]][,1:3],divergence=listDiv2[[i]][,1:4],xlow=0,xhigh=1,plot=F)
    r3 <- aMKT(daf=listDaf3[[i]][,1:3],divergence=listDiv3[[i]][,1:4],xlow=0,xhigh=1,plot=F)

    m1[i,1] <- r1$Fractions[3,1]
    m1[i,2] <- listDiv[[i]]$b
    m2[i,1] <- r2$Fractions[3,1]
    m2[i,2] <- listDiv[[i]]$b
    m3[i,1] <- r3$Fractions[3,1]
    m3[i,2] <- listDiv3[[i]]$b
}
```
